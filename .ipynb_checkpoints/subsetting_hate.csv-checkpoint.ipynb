```python
original_hate["Offense"] = original_hate['OFFENSE_NAME'].str.split(";").str[0].str.split(" ").str[-1]
```

```python
first3offense = list(original_hate["Offense"].value_counts()[0:3].index)
original_hate["Offense"][(original_hate["Offense"] != first3offense[0]) & 
              (original_hate["Offense"] != first3offense[1]) & 
              (original_hate["Offense"] != first3offense[2])] = "Others"
```

```python
import pandas as pd
df = pd.read_csv("hate_crime.csv", low_memory=False)
```

```python
df.head()
```

```python
df_small = df[['PUB_AGENCY_NAME', 'STATE_NAME']].rename(columns={'PUB_AGENCY_NAME':'city', 'STATE_NAME':'state'})
df_small.head()
```

```python
df_small.groupby('state').size().to_frame().reset_index()
```

```python
(df_small.groupby(['city', 'state']).size().to_frame().reset_index()
 .merge(df_small.groupby('state').size().to_frame(), on='state', how='outer')
 .rename(columns={'0_x': 'city_count', '0_y': 'state_count'})).plot.scatter(x='state', y='city_count', figsize=(20, 5), rot=90)

# (df_small.groupby(['city', 'state']).size().to_frame().reset_index()
#  .merge(df_small.groupby('state').size().to_frame(), on='state', how='outer')
#  .rename(columns={'0_x': 'city_count', '0_y': 'state_count'})).plot.scatter(x='state_count', y='city_count', figsize=(20, 5), rot=90)
```
